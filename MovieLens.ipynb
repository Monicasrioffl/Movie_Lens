                                     Q2.Can we find the movies that have more than one genre?

```python
#To find the information let us create a new column named Number of Genres and add it to the dataframe
movie["Number of Genres"]=movie.loc[:,genres].sum(axis=1) #it's a list now(num of genres)
#the new column adds the summation of one's from the genre columns along the row of a particular movie
#movie.loc[:,genre]-Picking up all the rows and columns that are present in the genres list
#axis=1--this parameter is used to get the summation of values along the columns
```

```python
movie[["movie id","Number of Genres"]]
```

```python
#we can print the head to see whether the new column has been added or not 
movie.head()
```

```python
#Let us pick up the rows where number of Genres>1 so we can easily find the num. of movies with >1
movie[movie['Number of Genres']>1] #this will subset the data as per the condition
```

                         Q3.Can we find the which movie has been  most preferred by the users?                                          

```python
#We can merge both the datasets on the basis of common variables that is movie_id
df_merge = movie.merge(ratings,on = 'movie id',how = 'inner')
df_merge.head()
```

```python
genres
```

```python
#Running the loop across the merged dataframes for the columns that are present in the genres list
for i in genres:
    print(i,':' ,df_merge[df_merge[i]==1].rating.mean())
```

INSIGHTS:
    'Film-Noir' genre is having the highest average rating followed by the 'War' genre
    'Fantasy' genre is having the lowest average ratings
    Among 18 genres only 5 genres has been received a rating below 3.5 which means the 72% of genres are rated above average
